```python
from bs4 import BeautifulSoup as bf
from urllib.request import urlopen
from prettytable import PrettyTable as pt
import pandas as pd
import numpy as np
import json
from rake_nltk import Rake
import time
from nltk.corpus import wordnet2021
```

```python
PREFIX = "https://www.prospects.ac.uk"
```

```python
def crawler(url):
    html = urlopen(url)
    obj = bf(html.read(), 'html.parser')
    url_list = []
    for i in obj.body.div.find_next_sibling("div").main.div.find_next_sibling("div").div.div.div.ul.children:
        url_list.append(PREFIX + i.a.attrs["href"])
    return url_list
```

```python
url = ["https://www.prospects.ac.uk/careers-advice/what-can-i-do-with-my-degree",
"https://www.prospects.ac.uk/careers-advice/applying-for-jobs",
"https://www.prospects.ac.uk/careers-advice/getting-a-job",
"https://www.prospects.ac.uk/careers-advice/interview-tips",
"https://www.prospects.ac.uk/careers-advice/cvs-and-cover-letters"
]
```

```python
url_list = []
for i in url:
    url_list.extend(crawler(i))
url_list = list(set(url_list))
```

```python
q_a_storge = []
```

```python
count = 0
for i in url_list:
    time.sleep(0.1)
    count += 1
    print(str(count) + "/" + str(len(url_list)))
    html = urlopen(i)
    obj = bf(html.read(), 'html.parser')
    q_a_list = []
    job_name = obj.body.div.find_next_sibling("div").main.article.div.div.div.div.div.header.h1.string
    pair = [job_name, ""]
    answer = ""
    for j in obj.body.div.find_next_sibling("div").main.article.section.div.div.div.find_next_sibling("div").div.children:
        if j.name == "footer": break
        if j.name == "h2":
            if j.string == "Find out more": break
            pair[1] = answer
            q_a_list.append(pair)
            pair = [j.string, ""]
            answer = ""
        elif j.name == "p":
            answer += "".join(j.strings)
            answer += "\n"
        elif j.name == "ul":
            for k in j.children:
                if k.name == "li":
                    answer += "* "
                    answer += "".join(k.strings)
                    answer += "\n"
        elif j.name == "figure":
            if "content-table" in j.attrs["class"]:
                table = j.div.table
                caption = "".join(j.figcaption.strings)
                thead = table.thead
                tbody = table.tbody
                field = []
                tb = pt()
                tb.align = "l"
                for th in thead.tr.children:
                    if th.name == "th":
                        field.append("".join(th.strings))
                tb.field_names = field
                for tr in tbody.children:
                    if tr.name == "tr":
                        row = []
                        for td in tr.children:
                            if td.name == "td":
                                row.append("".join(td.strings))
                        tb.add_row(row)
                answer += caption + "\n"
                answer += tb.get_string()
    pair[1] = answer
    q_a_list.append(pair)
    q_a_storge.append(q_a_list)
```

```python
q_list = []
for page in q_a_storge:
    for i in range(len(page)):
        if i != 0:
            q_list.append(page[i][0])
q_list = set(q_list)
```

```python
keyword_list = []
for q in q_list:
    r = Rake()
    r.extract_keywords_from_text(q)
    keyword_list.extend(r.get_ranked_phrases())
```

```python
keyword_list = set(keyword_list)
```

```python
keyword_sysnonym_list = []
for i in keyword_list:
    temp = []
    keyword = i.replace(" ", "_")
    for syn in wordnet2021.synsets(keyword):
        for l in syn.lemmas():
            temp.append(l.name().replace("_", " "))
    temp = list(set(temp))
    temp = [i] + temp
    if len(temp) > 1:
        keyword_sysnonym_list.append(temp)

print(keyword_sysnonym_list)
```

```python
synonym_output_list = []
for i in keyword_sysnonym_list:
    keyword = i.pop(0)
    string = '"' + keyword + '",'
    count = 0
    for j in i:
        if not j.isalnum(): continue
        string += '"' + j + '",'
        count += 1
        if count == 19:
            string = string[:-1]
            synonym_output_list.append(string)
            count = 0
            string = '"' + keyword + '",'
    string = string[:-1]
    synonym_output_list.append(string)
print(synonym_output_list)
```

```python
synonym_dataframe = pd.DataFrame({"Synonyms":[]})
for i in range(len(synonym_output_list)):
    synonym_dataframe.loc[i] = {"Synonyms": synonym_output_list[i]}
print(synonym_dataframe)
```

```python
synonym_dataframe.to_csv("synonyms.csv", index=False)
```
