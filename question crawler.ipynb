```python
from bs4 import BeautifulSoup as bf
from urllib.request import urlopen
from prettytable import PrettyTable as pt
import pandas as pd
import numpy as np
import json
import time
```

```python
PREFIX = "https://www.prospects.ac.uk"
```

```python
def crawler(url):
    html = urlopen(url)
    obj = bf(html.read(), 'html.parser')
    url_list = []
    for i in obj.body.div.find_next_sibling("div").main.div.find_next_sibling("div").div.div.div.ul.children:
        url_list.append(PREFIX + i.a.attrs["href"])
    return url_list
```

```python
url = "https://www.prospects.ac.uk/careers-advice/what-can-i-do-with-my-degree"
url_list = []
url_list.extend(crawler(url))
```

```python
url = ["https://www.prospects.ac.uk/careers-advice/applying-for-jobs",
"https://www.prospects.ac.uk/careers-advice/getting-a-job",
"https://www.prospects.ac.uk/careers-advice/interview-tips",
"https://www.prospects.ac.uk/careers-advice/cvs-and-cover-letters"
]
```

```python
url_list_2 = []
for i in url:
    url_list_2.extend(crawler(i))
url_list_2 = list(set(url_list_2))
url_list.extend(url_list_2)
```

```python
q_a_storge = []
```

```python
count = 0
for i in url_list:
    time.sleep(0.5)
    count += 1
    print(str(count) + "/" + str(len(url_list)))
    html = urlopen(i)
    obj = bf(html.read(), 'html.parser')
    q_a_list = []
    job_name = obj.body.div.find_next_sibling("div").main.article.div.div.div.div.div.header.h1.string
    pair = [job_name, ""]
    answer = ""
    for j in obj.body.div.find_next_sibling("div").main.article.section.div.div.div.find_next_sibling("div").div.children:
        if j.name == "footer": break
        if j.name == "h2":
            if j.string == "Find out more": break
            pair[1] = answer
            q_a_list.append(pair)
            pair = [j.string, ""]
            answer = ""
        elif j.name == "p":
            answer += "".join(j.strings)
            answer += "\n"
        elif j.name == "ul":
            for k in j.children:
                if k.name == "li":
                    answer += "* "
                    answer += "".join(k.strings)
                    answer += "\n"
        elif j.name == "figure":
            if "content-table" in j.attrs["class"]:
                table = j.div.table
                caption = "".join(j.figcaption.strings)
                thead = table.thead
                tbody = table.tbody
                field = []
                tb = pt()
                tb.align = "l"
                for th in thead.tr.children:
                    if th.name == "th":
                        field.append("".join(th.strings))
                tb.field_names = field
                for tr in tbody.children:
                    if tr.name == "tr":
                        row = []
                        for td in tr.children:
                            if td.name == "td":
                                row.append("".join(td.strings))
                        tb.add_row(row)
                answer += caption + "\n"
                answer += tb.get_string()
    pair[1] = answer
    q_a_list.append(pair)
    q_a_storge.append(q_a_list)
```

```python
qa_df = pd.DataFrame({'Question':[], 'Answer':[], 'Source':[], 'Metadata':[], "SuggestedQuestions":[], 'IsContextOnly':[], "Prompts":[], "QnaId":[]})
#qnald = 0
#print(np.array(q_a_storge).shape)
#print(q_a_storge)
print_wrong = True
qa_database = dict()
for page in q_a_storge:
    rec = qa_df.shape[0]
    prompt_dict_lst = []
    displayOrder = 0
    for i in range(len(page)):
        if i == 0:
            qa_df.loc[qa_df.shape[0]] = {'Question':page[0][0], 'Answer':page[0][0], 'Source': "Editorial", 'Metadata':'', "SuggestedQuestions":"[]", 'IsContextOnly':"FALSE", "Prompts":"", "QnaId":qa_df.shape[0] + 1}
            qa_database[page[0][0]] = page[i][1]
        else:
            qa_df.loc[qa_df.shape[0]] = {'Question':page[0][0] + " " + page[i][0], 'Answer':page[0][0] + " " + page[i][0], 'Source': "Editorial", 'Metadata':'', "SuggestedQuestions":"[]", 'IsContextOnly':"FALSE", "Prompts":"", "QnaId":qa_df.shape[0] + 1}
            qa_database[page[0][0] + " " +  page[i][0]] = page[i][1]
        if i == 0: continue
        prompt_dict = dict()
        prompt_dict['DisplayOrder'] = str(displayOrder)
        prompt_dict['QnaId'] = str(qa_df.shape[0])
        prompt_dict['DisplayText'] = page[i][0]
        prompt_dict_lst.append(json.dumps(prompt_dict))
        displayOrder += 1
    prompt = '['
    for i in prompt_dict_lst:
        prompt += i
        prompt += ','
    prompt = prompt[:-1] + "]"
    qa_df.iloc[rec, list(qa_df.columns).index("Prompts")] = prompt
    #res = json.loads(prompt)
    #print(qa_df)
job_q = ["Types of a job", "Responsibilities", "Salary", "Working hours", "What to expect", "Qualifications", "Skills", "Work experience", "Employers", "Professional development", "Career prospects"]
deg_q = ["Job options", "Work experience", "Employers", "Skills", "Further study", "What do the graduates do?"]
for i in list(set(job_q + deg_q)):
    qa_df.loc[qa_df.shape[0]] = {'Question':i, 'Answer':i, 'Source': "Editorial", 'Metadata':'', "SuggestedQuestions":"[]", 'IsContextOnly':"FALSE", "Prompts":"", "QnaId":qa_df.shape[0] + 1}
qa_df.to_csv("output.csv", index=False)
#qa_df.to_csv("output_small.csv", index=False)
```
